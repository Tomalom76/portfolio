# Pokazanie wykresu słupkowego wszystkich klas z podziałem na szerokość i długość płatka oraz kielicha #

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
df1.groupby('klasa (class)')['długość kielicha (sepal length)'].mean().plot(kind='bar', ax=axes[0, 0], color=['red', 'green', 'blue'])
axes[0, 0].set_title('Średnia długośc kielicha wg. rodzaju')
axes[0, 0].set_ylabel('Długośc Kielicha (cm)')

df1.groupby('klasa (class)')['szerokość kielicha (sepal width)'].mean().plot(kind='bar', ax=axes[0, 1], color=['red', 'green', 'blue'])
axes[0, 1].set_title('Średnia szerokośc kielicha wg. rodzaju')
axes[0, 1].set_ylabel('Szerokośc kielicha (cm)')

df1.groupby('klasa (class)')['długość płatka (petal length)'].mean().plot(kind='bar', ax=axes[1, 0], color=['red', 'green', 'blue'])
axes[1, 0].set_title('Średnia długośc płatka wg. rodzaju')
axes[1, 0].set_ylabel('Długośc płatka(cm)')

df1.groupby('klasa (class)')['szerokość płatka (petal width)'].mean().plot(kind='bar', ax=axes[1, 1], color=['red', 'green', 'blue'])
axes[1, 1].set_title('Średnia szerokośc płatka wg. rodzaju')
axes[1, 1].set_ylabel('Szerokośc płatka (cm)')

plt.tight_layout()
```

# Pokazanie wykresu punktowego według długość i szerokość płatka z uwzględnieniem podziału na klasy

```python
# Tworzenie wykresu punktowego
plt.figure(figsize=(10, 6))
for class_name, group in df1.groupby('klasa (class)'):
    plt.scatter(group['długość płatka (petal length)'], group['szerokość płatka (petal width)'], label=class_name)

plt.title('Wykres punktowy długości płatka vs szerokośc płatka z podziałem według rodzaju.')
plt.xlabel('Długośc Płatka (cm)')
plt.ylabel('Szerokośc Płatka (cm)')
plt.legend(title='Rodzaj')
plt.grid(True)
```

# Pokazanie wykresu typu boxplot z uwzględnieniem podziału na rodzaje, długości i szerokości płatków i kielichów.

```python
# Ustawienie palety kolorów
palette = {
    'Iris-setosa': 'red',
    'Iris-versicolor': 'green',
    'Iris-virginica': 'blue'
}

# Stworzenie wykresu
plt.figure(figsize=(12, 8))
sns.boxplot(x='variable', y='value', hue='klasa (class)', data=pd.melt(df1, id_vars='klasa (class)'), palette=palette)

# Ustawienie nazw labelów
plt.xlabel('Features')
plt.ylabel('Miara (cm)')
plt.title('Wykres typu boxplot z uwzględnieniem podziału na rodzaje')
```
